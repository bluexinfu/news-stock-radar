```python
import sys
from pathlib import Path
ROOT = Path("..").resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from src.analyzers.correlation import load_data, pearson_table, rolling_correlation_all
from src.analyzers.lead_lag import compute_lag_correlation, bootstrap_best_lag
from src.analyzers.event_study import run_event_study
from src.visualizers.plots import (
    plot_nii_vs_price, plot_rolling_correlation,
    plot_lag_correlation, plot_event_study, plot_correlation_heatmap,
    TICKER_NAMES, TICKER_COLORS,
)

TOPIC = "cowos"
REPORTS = ROOT / "reports"
REPORTS.mkdir(exist_ok=True)

aligned, nii_df = load_data(TOPIC)
nii = nii_df["nii"]

STOCK_TICKERS = {
    "2330.TW": "台積電",
    "3413.TW": "京鼎",
    "3583.TW": "辛耘",
    "3680.TWO": "家登",
    "3131.TWO": "弘塑",
}

print(f"對齊後資料：{aligned.index.min().date()} ~ {aligned.index.max().date()}  {len(aligned)} 交易日")
print(f"NII 統計：mean={nii.mean():.2f}  std={nii.std():.2f}  max={nii.max():.2f}")
```

## A1：NII 訊息強度 × 各標的股價 雙軸疊圖

**讀圖建議**：左軸藍色為 NII（值越高代表訊息量越大）；右軸橘紅色為股價。觀察兩條線的高低峰是否有時間上的對應關係。

```python
def safe_col(ticker):
    return "close_" + ticker.replace(".", "_").replace("^", "")

for ticker, name in STOCK_TICKERS.items():
    col = safe_col(ticker)
    if col not in aligned.columns:
        print(f"⚠️ 找不到 {col}，略過")
        continue
    plot_nii_vs_price(
        nii, aligned[col], ticker,
        save_path=REPORTS / f"a1_overlay_{ticker.replace('.','_')}.png"
    )
```

## A2：全期間皮爾森相關係數

使用日報酬率計算，避免共同趨勢造成的虛假相關。

**讀表建議**：r > 0.2 且 p < 0.05 才具參考意義；r > 0.3 視為有實質相關。

```python
pt = pearson_table(aligned, nii_df, nii_col="nii", use_returns=True)

# 加上中文名稱方便閱讀
name_map = {t.replace(".", "_"): n for t, n in TICKER_NAMES.items()}
pt["名稱"] = pt["ticker"].apply(lambda t: TICKER_NAMES.get(t, t))
display_cols = ["ticker", "名稱", "r", "p_value", "n", "significant"]
print("=== A2：NII × 各標的 全期間皮爾森相關係數（日報酬率）===")
print(pt[display_cols].to_string(index=False))
print()
print("注意：^TWII 為大盤，用於 A7 超額相關（本次不做）")
```

## A2b：NII × 概念股等權指數（SMI）整體相關

**為何需要 SMI？**
個股相關係數受單一公司特有事件（法說會、法人調節）干擾；等權籃子可平均掉個股雜訊，讓 NII 訊號的主題效應更清晰。

**SMI 定義**：
- 成份：CoWoS 概念股（排除 ^TWII 大盤）
- 方法：各股日報酬率等權平均 → 累積指數（基期=100）
- 相關計算：NII pct_change × SMI 日報酬率（均為「變化量」，避免虛假相關）

**讀表建議**：與 A2 個股比較，若 SMI r > 辛耘個股 r，代表等權籃子去噪後訊號更強。

```python
from src.processors.sector_index import build_smi, smi_summary
from scipy import stats as scipy_stats
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ── 建立 SMI ────────────────────────────────────────────────────────
smi_return = build_smi(TOPIC, aligned, save=True)
smi_level  = (1 + smi_return.fillna(0)).cumprod() * 100

stats = smi_summary(smi_return)
print("=== SMI 摘要統計 ===")
print(f"  成份數   : {len([c for c in aligned.columns if c.startswith('close_') and 'TWII' not in c])} 檔")
print(f"  年化波動 : {stats['ann_vol']:.1%}")
print(f"  年化報酬 : {stats['ann_ret']:.1%}")
print(f"  估算 Sharpe : {stats['sharpe_est']:.2f}")
print(f"  最大回撤 : {stats['max_dd']:.1%}")

# ── NII pct_change × SMI 日報酬率 相關 ──────────────────────────────
ret_nii = nii.pct_change()
ret_smi = smi_return
common  = ret_nii.dropna().index.intersection(ret_smi.dropna().index)

r_all, p_all = scipy_stats.pearsonr(ret_nii.loc[common], ret_smi.loc[common])
sig_all = "✓ 顯著" if p_all < 0.05 else "✗ 不顯著"

# 新聞覆蓋完整期（2025-07 後）
mask_recent = common >= pd.Timestamp("2025-07-01")
r_rec, p_rec = scipy_stats.pearsonr(
    ret_nii.loc[common[mask_recent]], ret_smi.loc[common[mask_recent]]
)
sig_rec = "✓ 顯著" if p_rec < 0.05 else "✗ 不顯著"

print(f"\n=== A2b：NII × SMI 相關係數 ===")
print(f"  全期      r={r_all:+.4f}  p={p_all:.4f}  n={len(common)}  {sig_all}")
print(f"  2025-07+  r={r_rec:+.4f}  p={p_rec:.4f}  n={mask_recent.sum()}  {sig_rec}")
print(f"\n  對照 A2 辛耘個股（全期）：r=+0.1493")
verdict = "▲ SMI 高於辛耘，等權籃子去噪有效" if abs(r_all) > 0.1493 else "▽ SMI 低於辛耘（個股辛耘信號最強）"
print(f"  {verdict}")

# ── A2b 視覺化：NII × SMI 雙軸疊圖 ─────────────────────────────────
fig, ax1 = plt.subplots(figsize=(14, 5))
ax1.fill_between(nii.index, nii, alpha=0.2, color="#1565C0")
ax1.plot(nii.index, nii, color="#1565C0", linewidth=1.8, label="NII 訊息強度")
ax1.set_ylabel("NII 訊息強度", color="#1565C0")
ax1.tick_params(axis="y", labelcolor="#1565C0")

ax2 = ax1.twinx()
ax2.plot(smi_level.index, smi_level, color="#6A1B9A", linewidth=2.0,
         label="CoWoS SMI（等權指數，基期=100）")
ax2.set_ylabel("SMI 指數水準", color="#6A1B9A")
ax2.tick_params(axis="y", labelcolor="#6A1B9A")

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left", fontsize=9)

ax1.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
plt.setp(ax1.xaxis.get_majorticklabels(), rotation=30)
plt.title(
    f"CoWoS 訊息強度（NII）× 概念股等權指數（SMI）雙軸疊圖\n"
    f"全期 r={r_all:+.4f}（p={p_all:.4f}）｜2025-07後 r={r_rec:+.4f}（p={p_rec:.4f}）",
    fontsize=11
)
fig.text(0.01, -0.02, "SMI=5 檔概念股等權指數（台積電/京鼎/辛耘/家登/弘塑）｜相關不等於因果", fontsize=8, color="gray")
plt.tight_layout()
save_path = REPORTS / "a2b_smi_overlay.png"
plt.savefig(save_path, bbox_inches="tight", dpi=150)
plt.show()
print(f"圖已存：{save_path}")
```

## A3：滾動相關係數（30天 / 60天窗口）

**讀圖建議**：觀察相關係數是否隨時間穩定，或只在特定事件期間才出現。穩定的正相關（r > 0.3）才代表訊號可靠。

```python
rolling_all = rolling_correlation_all(aligned, nii_df, windows=[30, 60])

for ticker in STOCK_TICKERS:
    if ticker not in rolling_all:
        continue
    plot_rolling_correlation(
        rolling_all[ticker], ticker,
        save_path=REPORTS / f"a3_rolling_{ticker.replace('.','_')}.png"
    )
```

## A4：Lead-Lag 分析

**讀圖建議**：若 r 在 lag > 0 時最高 → NII 領先股價（可能有預測意義）；若在 lag < 0 時最高 → 股價先動，訊息才跟上（追漲現象）。* 標記代表 p < 0.05。

```python
lag_results = {}
LAGS = list(range(-10, 11))

for ticker in STOCK_TICKERS:
    col = safe_col(ticker)
    if col not in aligned.columns:
        continue
    price = aligned[col]
    lag_df = compute_lag_correlation(nii, price, lags=LAGS)
    best = bootstrap_best_lag(nii, price, lags=LAGS, n_boot=500)
    lag_results[ticker] = {"lag_df": lag_df, "best": best}

    print(f"\n{ticker} ({TICKER_NAMES.get(ticker, ticker)})")
    print(f"  最佳 lag = {best['best_lag']}  r = {best['r_at_best_lag']:.4f}")
    print(f"  {best['interpretation']}")
    print(f"  95% Bootstrap CI：[{best['ci_lower']}, {best['ci_upper']}]")
    print(f"  p-value：{best['p_value']}")

    plot_lag_correlation(
        lag_df, ticker, best_lag_info=best,
        save_path=REPORTS / f"a4_lag_{ticker.replace('.','_')}.png"
    )
```

## A5：事件研究 — NII 峰值日前後 ±10 天股價變化

**讀圖建議**：藍色粗線為平均累積報酬率（CAR）；淡藍陰影為 ±1 標準差。若事件日後（x > 0）有系統性正報酬，代表訊息爆發後股價確實上漲。

```python
event_days, _ = run_event_study(nii, aligned[safe_col("2330.TW")], k=1.5, min_gap=20, window=10)
print(f"識別出 {len(event_days)} 個 NII 峰值事件日：")
for d in sorted(event_days):
    print(f"  {d.date()}  NII={nii.loc[d]:.2f}")

for ticker in STOCK_TICKERS:
    col = safe_col(ticker)
    if col not in aligned.columns:
        continue
    ev_days, car_df = run_event_study(nii, aligned[col], k=1.5, min_gap=20, window=10)
    if car_df.empty:
        print(f"⚠️ {ticker} 事件研究資料不足，略過")
        continue
    plot_event_study(
        car_df, ev_days, ticker,
        save_path=REPORTS / f"a5_event_{ticker.replace('.','_')}.png"
    )
```

```python
print("=" * 65)
print("Phase 1 CoWoS 分析結論摘要")
print("=" * 65)

print("\n【A2 全期間相關係數（日報酬率）】")
for _, row in pt_stocks.iterrows():
    sig = "✓ 顯著" if row["significant"] else "✗ 不顯著"
    print(f"  {row['ticker']:12s} {row.get('名稱',''):6s}  r={row['r']:+.4f}  p={row['p_value']:.4f}  {sig}")

print("\n【A4 最佳 Lead-Lag】")
for ticker, res in lag_results.items():
    b = res["best"]
    print(f"  {ticker:12s}  lag={b['best_lag']:+3d}天  r={b['r_at_best_lag']:+.4f}  {b['interpretation']}")

print()
print("【Phase 1 成功標準對照】")
best_r = pt_stocks["r"].abs().max()
print(f"  最強相關係數：{best_r:.3f}")
if best_r >= 0.3:
    print("  ✅ 有觀察到清晰訊號（|r| > 0.3），可考慮進入 Phase 2")
elif best_r >= 0.1:
    print("  ⚠️ 訊號模糊（0.1 < |r| < 0.3），建議迭代方法論後再決定")
else:
    print("  ❌ 無明顯訊號（|r| < 0.1），考慮換主題重做")

print()
print("注意：上述相關係數以日報酬率計算（去趨勢），比原始價格的結果更保守且更可信")
```
