```python
import sys
from pathlib import Path
ROOT = Path("..").resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from src.analyzers.correlation import load_data, pearson_table, rolling_correlation

plt.rcParams["font.family"] = ["Arial Unicode MS", "PingFang TC", "sans-serif"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["figure.dpi"] = 120

TOPIC = "cowos"
REPORTS = ROOT / "reports"

aligned, nii_df = load_data(TOPIC)

STOCK_TICKERS = {
    "2330.TW":  "台積電",
    "3413.TW":  "京鼎",
    "3583.TW":  "辛耘",
    "3680.TWO": "家登",
    "3131.TWO": "弘塑",
}

def safe_col(ticker):
    return "close_" + ticker.replace(".", "_").replace("^", "")

def pearson_r(s1, s2):
    common = s1.index.intersection(s2.index)
    a, b = s1.loc[common].dropna(), s2.loc[common].dropna()
    c = a.index.intersection(b.index)
    r, p = stats.pearsonr(a.loc[c], b.loc[c])
    return round(r, 4), round(p, 5), len(c)

print("資料載入完成，開始穩健性檢驗...")
```

## R1：NII 權重敏感度分析

改變 Trends 與新聞數的權重比例，看相關係數結論是否改變。

- `w30`：Trends 30%，新聞數 70%（偏重新聞）
- `w50`：各 50%（預設）
- `w70`：Trends 70%，新聞數 30%（偏重搜尋熱度）

```python
weight_cols = {"w30": "nii_w30", "w50": "nii_w50", "w70": "nii_w70"}
weight_results = {w: {} for w in weight_cols}

for w_label, nii_col in weight_cols.items():
    nii_sig = nii_df[nii_col].diff()
    for ticker in STOCK_TICKERS:
        col = safe_col(ticker)
        if col not in aligned.columns:
            continue
        price_ret = aligned[col].pct_change()
        r, p, n = pearson_r(nii_sig, price_ret)
        weight_results[w_label][ticker] = {"r": r, "p": p}

# 整理成矩陣顯示
idx = list(STOCK_TICKERS.keys())
r_matrix = pd.DataFrame(
    {w: {t: weight_results[w][t]["r"] for t in idx} for w in weight_cols}
)
p_matrix = pd.DataFrame(
    {w: {t: weight_results[w][t]["p"] for t in idx} for w in weight_cols}
)
r_matrix.index = [f"{t}({n})" for t, n in STOCK_TICKERS.items()]
p_matrix.index = r_matrix.index

print("=== R1：三組權重的皮爾森 r（日報酬率）===")
print(r_matrix.to_string())
print()
print("=== p-value ===")
print(p_matrix.to_string())

# 視覺化
fig, ax = plt.subplots(figsize=(9, 4))
x = np.arange(len(idx))
width = 0.25
colors = ["#90CAF9", "#1565C0", "#0D47A1"]
for i, (w, col) in enumerate(weight_cols.items()):
    vals = [weight_results[w][t]["r"] for t in idx]
    bars = ax.bar(x + i * width, vals, width, label=f"NII {w}", color=colors[i], alpha=0.85)

ax.set_xticks(x + width)
ax.set_xticklabels([f"{t}\n{n}" for t, n in STOCK_TICKERS.items()], fontsize=9)
ax.axhline(0, color="gray", linewidth=0.8)
ax.axhline(0.1, color="green", linewidth=0.6, linestyle=":", alpha=0.6)
ax.set_ylabel("皮爾森 r（日報酬率）")
ax.set_title("R1：NII 權重敏感度 — 三組權重的相關係數比較", fontsize=12)
ax.legend(fontsize=9)
plt.tight_layout()
plt.savefig(REPORTS / "r1_weight_sensitivity.png", bbox_inches="tight", dpi=150)
plt.show()
```

## R2：只用 Trends（排除新聞數干擾）

Google News RSS 只有 12 個月歷史，GDELT CoWoS 關鍵字也有缺漏，新聞數資料在 2024 年較稀疏。
若純用 Trends（完整 2 年），結論是否更清晰？

```python
trends_only = aligned["trends_interp"].diff()

print("=== R2：只用 Google Trends（純搜尋熱度）===")
print(f"{'標的':12s} {'名稱':6s} {'r':>8s} {'p':>9s} {'n':>5s}")
print("-" * 45)

trends_only_results = {}
for ticker, name in STOCK_TICKERS.items():
    col = safe_col(ticker)
    if col not in aligned.columns:
        continue
    price_ret = aligned[col].pct_change()
    r, p, n = pearson_r(trends_only, price_ret)
    sig = "✓" if p < 0.05 else "✗"
    print(f"{ticker:12s} {name:6s} {r:>+8.4f} {p:>9.5f}  {n}  {sig}")
    trends_only_results[ticker] = {"r": r, "p": p}

print()
print("=== 比較：只用 Trends vs NII(w=0.5) ===")
print(f"{'標的':12s} {'名稱':6s} {'r_trends_only':>14s} {'r_nii_w50':>10s} {'差異':>8s}")
print("-" * 52)
nii_sig = nii_df["nii_w50"].diff()
for ticker, name in STOCK_TICKERS.items():
    col = safe_col(ticker)
    if col not in aligned.columns:
        continue
    price_ret = aligned[col].pct_change()
    r_nii, _, _ = pearson_r(nii_sig, price_ret)
    r_t = trends_only_results[ticker]["r"]
    print(f"{ticker:12s} {name:6s} {r_t:>+14.4f} {r_nii:>+10.4f} {r_t - r_nii:>+8.4f}")
```

## R3：時間段切割

全期（2024-05~2026-05）vs 新聞覆蓋完整期（2025-07~2026-05）。
若 2025-07 後的 r 明顯更高，代表資料品質影響了主分析結論。

```python
periods = {
    "全期 (2024-05~2026-05)": (None, None),
    "早期 (2024-05~2025-06)": (None, "2025-07-01"),
    "新聞完整期 (2025-07~2026-05)": ("2025-07-01", None),
}

nii_sig_full = nii_df["nii"].diff()

period_results = {p: {} for p in periods}
for period_label, (start, end) in periods.items():
    mask = pd.Series(True, index=nii_sig_full.index)
    if start:
        mask &= nii_sig_full.index >= pd.Timestamp(start)
    if end:
        mask &= nii_sig_full.index < pd.Timestamp(end)
    sig_sub = nii_sig_full[mask]

    for ticker, name in STOCK_TICKERS.items():
        col = safe_col(ticker)
        if col not in aligned.columns:
            continue
        price_ret = aligned[col].pct_change()[mask]
        r, p, n = pearson_r(sig_sub, price_ret)
        period_results[period_label][ticker] = {"r": r, "p": p, "n": n}

# 顯示
for period_label in periods:
    print(f"\n=== {period_label} ===")
    for ticker, name in STOCK_TICKERS.items():
        if ticker not in period_results[period_label]:
            continue
        d = period_results[period_label][ticker]
        sig = "✓" if d["p"] < 0.05 else "✗"
        print(f"  {ticker:12s} {name:6s}  r={d['r']:+.4f}  p={d['p']:.4f}  n={d['n']}  {sig}")

# 視覺化：三期 × 五標的
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
colors = ["#1565C0", "#42A5F5", "#90CAF9"]

for ax, (period_label, _) in zip(axes, periods.items()):
    tickers = list(STOCK_TICKERS.keys())
    names = [STOCK_TICKERS[t] for t in tickers]
    rs = [period_results[period_label].get(t, {}).get("r", 0) for t in tickers]
    ps = [period_results[period_label].get(t, {}).get("p", 1) for t in tickers]
    bar_colors = ["#1565C0" if p < 0.05 else "#BDBDBD" for p in ps]
    ax.barh(names, rs, color=bar_colors, alpha=0.85)
    ax.axvline(0, color="gray", linewidth=0.8)
    ax.axvline(0.1, color="green", linewidth=0.6, linestyle=":", alpha=0.6)
    ax.set_title(period_label.split("(")[0].strip(), fontsize=10)
    ax.set_xlabel("r")

axes[0].set_ylabel("標的")
fig.suptitle("R3：時間段切割 — 相關係數比較（藍色=p<0.05，灰色=不顯著）", fontsize=12)
plt.tight_layout()
plt.savefig(REPORTS / "r3_time_period.png", bbox_inches="tight", dpi=150)
plt.show()
```

## R4：滾動窗口大小（30 / 60 / 90 天）

不同的滾動窗口是否改變結論？窗口越大越平滑、越小越敏感。

```python
# 以辛耘（訊號最強）為代表展示三種窗口
ticker = "3583.TW"
col = safe_col(ticker)
roll_df = rolling_correlation(aligned, nii_df, col, nii_col="nii", windows=[30, 60, 90])

fig, ax = plt.subplots(figsize=(14, 4))
palette = {"rolling_corr_30d": "#0D47A1", "rolling_corr_60d": "#1976D2", "rolling_corr_90d": "#90CAF9"}
labels = {"rolling_corr_30d": "30天窗口", "rolling_corr_60d": "60天窗口", "rolling_corr_90d": "90天窗口"}

for col_name, color in palette.items():
    ax.plot(roll_df.index, roll_df[col_name], color=color, linewidth=1.5, label=labels[col_name])

ax.axhline(0, color="gray", linewidth=0.8)
ax.axhline(0.2, color="green", linewidth=0.6, linestyle=":", alpha=0.5)
ax.axhline(-0.2, color="red", linewidth=0.6, linestyle=":", alpha=0.5)
ax.set_ylabel("滾動相關係數 r")
ax.set_ylim(-0.8, 0.8)
ax.legend(fontsize=9)
import matplotlib.dates as mdates
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=30)
ax.set_title("R4：NII × 辛耘（3583.TW）滾動相關 — 三種窗口比較（日報酬率）", fontsize=12)
plt.tight_layout()
plt.savefig(REPORTS / "r4_window_size.png", bbox_inches="tight", dpi=150)
plt.show()

# 各窗口的平均 r（整體訊號強度摘要）
print("各窗口的平均滾動相關 r（辛耘）：")
for c, lbl in labels.items():
    mean_r = roll_df[c].dropna().mean()
    pct_pos = (roll_df[c].dropna() > 0.1).mean() * 100
    print(f"  {lbl}：平均 r={mean_r:+.4f}，r>0.1 佔 {pct_pos:.1f}% 時間")
```

## R5：週報酬率 vs 日報酬率

Trends 本身是週頻資料，日頻內插可能引入雜訊。改用週報酬率是否讓訊號更清晰？

```python
# 週報酬率：用每週最後一個交易日的收盤價計算
nii_weekly = nii_df["nii"].resample("W-FRI").last().diff()

print("=== R5：週報酬率 vs 日報酬率 ===")
print(f"{'標的':12s} {'名稱':6s} {'r_日':>9s} {'r_週':>9s} {'訊號增強?':>10s}")
print("-" * 52)

for ticker, name in STOCK_TICKERS.items():
    col = safe_col(ticker)
    if col not in aligned.columns:
        continue

    # 日報酬率
    price_daily = aligned[col].pct_change()
    nii_daily = nii_df["nii"].diff()
    r_d, _, _ = pearson_r(nii_daily, price_daily)

    # 週報酬率
    price_weekly = aligned[col].resample("W-FRI").last().pct_change()
    r_w, p_w, n_w = pearson_r(nii_weekly, price_weekly)
    sig = "✓" if p_w < 0.05 else "✗"
    better = "↑ 增強" if abs(r_w) > abs(r_d) else "↓ 減弱"
    print(f"{ticker:12s} {name:6s} {r_d:>+9.4f} {r_w:>+9.4f}  {better} {sig}")

# 視覺化：週 vs 日的相關係數比較
fig, ax = plt.subplots(figsize=(10, 4))
tickers = list(STOCK_TICKERS.keys())
names = [STOCK_TICKERS[t] for t in tickers]
x = np.arange(len(tickers))
w = 0.35

r_daily_list, r_weekly_list = [], []
for ticker in tickers:
    col = safe_col(ticker)
    if col not in aligned.columns:
        r_daily_list.append(0); r_weekly_list.append(0)
        continue
    nii_d = nii_df["nii"].diff()
    price_d = aligned[col].pct_change()
    rd, _, _ = pearson_r(nii_d, price_d)
    price_w = aligned[col].resample("W-FRI").last().pct_change()
    rw, _, _ = pearson_r(nii_weekly, price_w)
    r_daily_list.append(rd)
    r_weekly_list.append(rw)

ax.bar(x - w/2, r_daily_list, w, label="日報酬率", color="#1565C0", alpha=0.85)
ax.bar(x + w/2, r_weekly_list, w, label="週報酬率", color="#42A5F5", alpha=0.85)
ax.set_xticks(x)
ax.set_xticklabels([f"{t}\n{n}" for t, n in STOCK_TICKERS.items()], fontsize=9)
ax.axhline(0, color="gray", linewidth=0.8)
ax.set_ylabel("皮爾森 r")
ax.legend()
ax.set_title("R5：日報酬率 vs 週報酬率的相關係數比較", fontsize=12)
plt.tight_layout()
plt.savefig(REPORTS / "r5_return_frequency.png", bbox_inches="tight", dpi=150)
plt.show()
```

# 穩健性檢驗（Robustness Check）

**目的**：確認 Phase 1 主分析的結論是否因方法論選擇而改變。

檢驗維度：
1. **NII 權重敏感度**：w_trends ∈ {0.3, 0.5, 0.7} — 改變 Trends 與新聞數的比重
2. **只用 Trends（純搜尋熱度）**：排除新聞數覆蓋不足的干擾
3. **時間段切割**：全期 vs 2025-07 後（新聞覆蓋較完整）
4. **滾動窗口大小**：30天、60天、90天
5. **報酬率頻率**：日報酬率 vs 週報酬率
